```python
def get_city_match_count(city):
  mask=df['city']==city
  return df[mask].shape[0]
```

```python
get_city_match_count('Pune')
```

```python
mask1 = df['city'] == 'Hyderabad'
mask2 = df['date'] > '2009-01-01'
df[mask1 & mask2]
```

```python
mask1 = df['team1'] == 'Sunrisers Hyderabad'
mask2 = df['team2'] =='Delhi Daredevils'
mask3 = df['winner'] == 'Delhi Daredevils'
df[mask1 & mask2 & mask3]
```

```python
# Filter the DataFrame to show only rows where 'winner' is 'Sunrisers Hyderabad'
sunrisers_wins = df[df['winner'] == 'Sunrisers Hyderabad']

# Get the number of matches won by Sunrisers Hyderabad
num_sunrisers_wins = sunrisers_wins.shape[0]

print(f"Number of matches won by Sunrisers Hyderabad: {num_sunrisers_wins}")
```
